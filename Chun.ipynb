**Numerical Features - Analysics**

```python
def numerical_feature_analysis(df,col):
  plt.figure(figsize=(15,6))
  sns.histplot(data=df,x=col,kde=True)
  plt.title(f"Distribution of {col} ")

  #Calculte the mean and median
  col_mean = df[col].mean()
  col_median = df[col].median()


  #Add vertical lines for mean and median
  plt.axvline(col_mean,color='red',linestyle='dashed',linewidth=2,label=f'Mean:{col_mean:.2f}')
  plt.axvline(col_median,color='green',linestyle='-',linewidth=2,label=f'Median:{col_median:.2f}')
  plt.legend()
  plt.tight_layout()
  plt.grid(True)
  plt.show()
  print()
  print()
```

```python
plot1 = ["tenure","MonthlyCharges","TotalCharges"]
for col in plot1:
  numerical_feature_analysis(df,col)
```

```python
def plot_boxplot(df,col):
  plt.figure(figsize=(10,6))
  sns.boxplot(y=df[col])
  plt.title(f"Boxplot of {col}")
  plt.ylabel(col)
  plt.show()
  print()
```

```python
plot2 = ["tenure","MonthlyCharges","TotalCharges"]
for col in plot2:
   plot_boxplot(df,col)
```
