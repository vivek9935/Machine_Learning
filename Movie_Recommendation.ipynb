```python
import numpy as np
import pandas as pd
import difflib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
df = pd.read_excel('/content/movies.xlsx',)
```

```python
selected_feature = ['genres','keywords','tagline','cast','director']
print(selected_feature)
```

```python
#Replacing the null value with null string
for feature in selected_feature:
  df[feature] = df[feature].fillna('')
```

```python
#combining all the 5 selected features
combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
combined_features
```

```python
#converting the text data into feature factor
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#getting the similarty score using cosince similarty
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
#getting the movie name from usier
movie_name = input('Enter your favourite movie name : ')
```

```python
#Creating the list of movie in the dataset
#Creating the list of movie in the dataset
list_of_all_titles = df['title'].astype(str).tolist()#cleate the list of the movie and converting it to string type.
print(list_of_all_titles)
```

```python
#Finding the close match for the movue
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie =int( df[df.title == close_match]['index'].iloc[0]) # Use .iloc[0] to access the first element of the Series
```

```python
print(index_of_the_movie)
```

```python
#getting a list of similar movie
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#Sorting the value based on similarty score
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```
